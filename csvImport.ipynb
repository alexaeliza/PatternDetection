```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from pathlib import Path
```

```python
def plotGivenFile(file, field='vw'):
    print(file)
    data = pd.read_csv(file)[field]
    plt.plot(data)
    plt.show()
```

```python
def plotAllFiles():
    directory = os.path.join(Path(os.getcwd()).parent.absolute(), "Resources")
    for subDirectory in os.listdir(directory):
        if not subDirectory.startswith('.'):
            for file in os.listdir(os.path.join(directory, subDirectory)):
                if not file.startswith('.'):
                    plotGivenFile(os.path.join(directory, subDirectory, file), 'h')
```

```python
plotAllFiles()
```
